Now that we know that the data is uniforam and is paired with its appropriate parents. I had a thought for how I will tackle the String database problem. I have realized that on Ensemble I can download the PEP file for HomoSapiens. I did that and can now match ProteinIDs with Gene Names which are used in most of the other files. 


```python
# Dict to store data
ensembl_dict = dict()
with gzip.open('/Users/andrewpowers/bioinformatics/parisa_project/Network-based-study-of-the-role-of-host-pathogen-interactions/database_files/ensembl/Homo_sapiens.GRCh38.pep.all.fa.gz', 'rt') as string_links:
    # Iter through ever line
    for line in string_links:
        # Only return the lines that have the > in the line
        if '>' in line:
            # Strip the newline character and create a list that contains each element in it
            first = line.strip('\n').split(' ')
            # Pull out the protein id
            protein_id = first[0].strip('>').split('.')[0]
            
            if len(first) < 8:
                gene_name = first[3].split(':')[1].split('.')[0]
            else:
                # Pull out the gene_name
                gene_name = first[7].split(':')[1]
            
            # Create a new entry with protein_id as key, and gene_name as the value
            ensembl_dict[protein_id] = gene_name


    
```

```python
ensembl_dict
```

Now we have a dict of Protein_IDs with corresponding gene symbol values. Now I can add the protein_id and the gene_name so I can combine things.

```python
string_data = pd.read_table('/Users/andrewpowers/bioinformatics/parisa_project/Network-based-study-of-the-role-of-host-pathogen-interactions/database_files/string/9606.protein.links.full.v11.5.txt.gz', delimiter=' ')
string_data.protein1 = string_data.protein1.map(lambda x: x.lstrip('9606.'))
string_data.protein2 = string_data.protein2.map(lambda x: x.lstrip('9606.'))
string_data.head()
```

Now that we have the string interaction link file imported correctly and removed the `9606.` from the beginning of the protein names. We can now use the protein IDs to supply the gene names to them. Using the ensembl_dict that I created above.

```python
# Generate the genes associated with the protein_id
string_data['gene1'] = [ensembl_dict[x] if x in ensembl_dict else "None" for x in string_data.protein1]
string_data['gene2'] = [ensembl_dict[x] if x in ensembl_dict else "None" for x in string_data.protein2]
```

Lets calculate the amount of NAs that are currently in the dataset now from what just happened.

```python
print('Gene1 total NA:', sum(string_data['gene1'] == 'None'), sep='\t')
print('Gene2 total NA:', sum(string_data['gene2'] == 'None'), sep='\t')
print('Gene1 NA proportion:', sum(string_data['gene1'] == 'None')/len(string_data), sep='\t')
print('Gene2 NA proportion:', sum(string_data['gene2'] == 'None')/len(string_data), sep='\t')
```
